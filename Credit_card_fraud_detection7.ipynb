# CREATING AND ASSIGNING DYNAMIC BINS FOR NUMERICAL DATA


```python
# Create a new column for bins
df['Amount-Bins'] = ''

# Function to create bins and labels
def make_bins(predictor, size=50):
    """
    Takes the predictor (a series or a dataframe of a single predictor) and size of bins
    Returns bins and bin labels
    """
    bins = np.linspace(predictor.min(), predictor.max(), num=size)
    bin_labels = []

    # Create bin labels
    for i in range(len(bins) - 1):
        bin_labels.append(f"{bins[i]:.2f}-{bins[i+1]:.2f}")

    return bins, bin_labels

# Apply the function to create bins
bins, bin_labels = make_bins(df['Amount'])

# Assign bins to the 'Amount-Bins' column
df['Amount-Bins'] = pd.cut(df['Amount'], bins=bins, labels=bin_labels, include_lowest=True)

# Display the first few rows to verify the bins
print(df[['Amount', 'Amount-Bins']].head())
```

# COMPARATIVE ANALYSIS OF CLASSIFICATION MODELS: LOGISTIC REGRESSION, NAIVE BAYES, AND ANOMALY DETECTION

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.metrics import classification_report, accuracy_score

# Initialize the models
log_reg = LogisticRegression()
naive_bayes = GaussianNB()
isolation_forest = IsolationForest(random_state=42)
local_outlier_factor = LocalOutlierFactor()

# Train and evaluate Logistic Regression
log_reg.fit(xtrain, ytrain)
y_pred_log_reg = log_reg.predict(xtest)
print("Logistic Regression:")
print(classification_report(ytest, y_pred_log_reg))
print("Accuracy:", accuracy_score(ytest, y_pred_log_reg))

# Train and evaluate Naive Bayes
naive_bayes.fit(xtrain, ytrain)
y_pred_naive_bayes = naive_bayes.predict(xtest)
print("Naive Bayes:")
print(classification_report(ytest, y_pred_naive_bayes))
print("Accuracy:", accuracy_score(ytest, y_pred_naive_bayes))

# Train and evaluate Isolation Forest
isolation_forest.fit(xtrain)
y_pred_iso_forest = isolation_forest.predict(xtest)
y_pred_iso_forest = [1 if x == -1 else 0 for x in y_pred_iso_forest]  # Convert -1 to 1 (fraud) and 1 to 0 (normal)
print("Isolation Forest:")
print(classification_report(ytest, y_pred_iso_forest))
print("Accuracy:", accuracy_score(ytest, y_pred_iso_forest))

# Train and evaluate Local Outlier Factor
y_pred_lof = local_outlier_factor.fit_predict(xtest)
y_pred_lof = [1 if x == -1 else 0 for x in y_pred_lof]  # Convert -1 to 1 (fraud) and 1 to 0 (normal)
print("Local Outlier Factor:")
print(classification_report(ytest, y_pred_lof))
print("Accuracy:", accuracy_score(ytest, y_pred_lof))
```

# GAUSSIAN NAIVE BAYES MODEL: PERFORMANCE EVALUATION AND ROC CURVE ANALYSIS

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, accuracy_score, roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Initialize the Gaussian Naive Bayes model
nb = GaussianNB()

# Train the model on the training data
nb.fit(xtrain, ytrain)

# Make predictions on the testing data
nb_pred = nb.predict(xtest)

# Evaluate the model using a classification report
print("Classification Report:")
print(classification_report(ytest, nb_pred))

# Calculate accuracy
accuracy = accuracy_score(ytest, nb_pred)
print("Accuracy:", accuracy)

# Calculate ROC curve and AUC
y_pred_prob = nb.predict_proba(xtest)[:, 1]
fpr, tpr, thresholds = roc_curve(ytest, y_pred_prob)
roc_auc = roc_auc_score(ytest, y_pred_prob)

# Plot the ROC curve with custom colors
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='green', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')  # Green ROC curve
plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--', label='Random Guess')  # Red diagonal line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
